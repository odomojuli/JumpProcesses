###Binomial Distribution

Theorem: Let $b(k;n,p)$ be the probability that $n$ Bernoulli trials with probabilities $p$ for success and $q=1-p$ 
for failure result in $k$ successes and $n-k$ failures. Then: 

\begin{equation}
b(k;n,p)={n \choose k} p^{k}q^{n-k}
\end{equation}

In particular, the probability of no success is $q^n$ and the probability of at least one success is $(1-q)^n$.

**CLT**: When $p \approx \frac{1}{2}$, as $n \rightarrow \infty$, the binomial distribution converges to the normal. In continuous time we are talking about Brownian motions. 

What happens to Brownian motion as time shrinks? The probability of moving up or down remains $p=\frac{1}{2}$ but the size of the movement reduces (the variance scales with time). 



###Poisson Limit Theorem

If $n$ is large and $p$ is small (rare event) the product $\lambda=np$ (mean of the r.v.) is of moderate size, we get a Poisson approximation for different values of $k$.

For $k=0$, prob=$(1-p)^{n}=(1-\frac{\lambda }{n})^{n}\approx e^{-\lambda }
$ as $n\rightarrow \infty $.

To see this, take logs:

\begin{equation}
\ln(\mbox{prob})=n\ln (1-\frac{\lambda }{n})\approx -\lambda -\frac{\lambda ^{2}}{2n})
\end{equation}

because $n\ln (1-\frac{\lambda }{n}) \approx \ln (e^{-\frac{\lambda }{n}})$. 

For $k=1$, prob=${n \choose 1} p^{1}(1-p)^{n-1}=np (1-p)^{n-1} \approx \lambda e^{-\lambda }$ when $p$ is very small.

For $k=2$, prob=${n \choose 2} p^{2}(1-p)^{n-2}=\frac{n(n-1)}{2!} p^{2}(1-p)^{n-2} \approx \frac{1}{2} \lambda^{2} e^{-\lambda }=\frac{\lambda ^{2}e^{-\lambda }}{2!}$.

Thus we have a Poisson distribution, where, for $n$ large, $p$ small, $\lambda = np$,  

\begin{equation}
b(k;n,p)\approx \frac{\lambda ^k}{k!}e^{-\lambda }
\end{equation}


###Poisson Processes

Also known as Jump Processes (Point Processes). For a detailed reference see "Point Processes and Queues" by P. Bremand. Generally, see Levy processes. 

**Pure jump process**

This is written as

\begin{equation}
dS=dQ(\lambda )
\end{equation}

which means

\begin{eqnarray*}
dQ(t) &=& 1 \mbox{   with prob   } \lambda\; dt \\
dQ(t) &=& 0 \mbox {   with prob   } 1-\lambda \; dt
\end{eqnarray*}

Note that the size of $dQ$ remains at 1 or 0 irrespective of the time interval. That is, as time shrinks, the size of the movement remains unchanged, but the probability of $dQ=1$ shrinks with time. This is the opposite of what we see with Brownian motion. Note that

\begin{equation}
E[dQ] = \lambda \; dt
\end{equation}

Suppose, for constant $J$ 

\begin{equation}
dS=J\; dQ(\lambda )
\end{equation}

Define a contingent claim $F(S,t)$. If the jump occurs, the change in the value of F, caused by the jump will be

\begin{equation}
\left[ F(S+J,t)-F(S,t)\right]
\end{equation}

The expected change is $\lambda \left[ F(S+J,t)-F(S,t)\right] dt$.

**Technical notes**

a) Let T$_{1,}T_{2,} \dots$ be iid random variables with exponential distribution. $P(T_j\in dt)=\lambda e^{-\lambda\; dt}dt$, $t>0$; $S_0=0,S_n=\sum\limits_{j=1}^nT_j$ for $n \geq 1$. $T_j^{\prime }$s are the inter-arrival times. $S_n$ is the arrival time of the $n^{th}$ item.

- $Q_t$ is a counting process. The number of counts in time $t = \#\{n \geq 1;S_n\leq t\}$ for all $t<\infty $. 
- $Q_t$ is the number of arrivals. It is a Poisson process with intensity $\lambda >0$.
- $Q_t$ is a step, or staircase, function.

b) Properties of $Q_t$

1. Increments $Q_{T_1}$, $Q_{T_2}$, $Q_{T_3}$, $Q_{T_4}$, $Q_{T_5}$, are independent.
2. $Q_\theta -Q_t$ is Poisson with parameter $\lambda (\theta -t)$,
therefore 
\begin{equation}
P(Q_\theta - Q_t = k) = \frac 1{k!}e^{-\lambda (\theta
-t)}\lambda (\theta -t)^k, \forall k=0, 1, 2 , 3, ...
\end{equation}
3. We can define the **compensated process** which is a martingale: $
\overline{Q}_t=Q_t-\lambda t$ for $0 \leq t\leq \infty $.


###Ito's Lemma Extended

Given an arithmetic jump-diffusion model (arithmetic Brownian motion with jumps)

\begin{equation}
dS = \alpha \; dt + \sigma \; dZ + J \; dQ(\lambda)
\end{equation}

Given a function $F(S,t)$, 

\begin{equation}
dF=\frac 12F_{ss}(dS)^2+F_sdS+F_tdt+\left[ F(S+J,t)-F(S,t)\right] dQ 
\end{equation}

The diffusion moves show a differential, then the jump move is like a difference, therefore

\begin{eqnarray*}
dF &=& \frac 12\frac{\partial ^2F}{\partial S^2}\sigma ^2dt+\frac{\partial F}{
\partial S}\left[ \alpha dt+\sigma dZ\right] +\frac{\partial F}{\partial t} dt
+\left[ F(S+J,t)-F(S,t)\right] dQ 
\end{eqnarray*}

##Contingent Claim Pricing

In the risk-neutral world, the expected return on the contingent clam must be equal to the risk-free rate. 

\begin{equation}
E\left[ dF\right] =\left[ r+\pi (S,t,\lambda )\right] F\; dt 
%0 &=&\frac 12F_{ss}\sigma ^2+F_sdS+F_tdt+\left[ F(S+J,t)-F(S,t)\right] dQ
\end{equation}

If jump risk is diversifiable, then $\pi =0$ (where this is the risk premium for undiversifiable jump risk). 

**Example**

We rederive the Black-Scholes PDE using the equation above. Given that there are no jumps in the BS world, the change in $F$ in the risk-neutral world, where $\alpha = r$, is:

\begin{equation}
dF = \frac 12\frac{\partial ^2F}{\partial S^2}\sigma ^2dt+\frac{\partial F}{
\partial S}\left[ r dt+\sigma dZ\right] +\frac{\partial F}{\partial t} dt 
\end{equation}

Noting that $\pi=0$ as there is no jump risk, we have

\begin{equation}
E(dF) = r F \;dt
\end{equation}

Substituting for $dF$, we have the BS PDE:

\begin{equation}
\frac 12\frac{\partial ^2F}{\partial S^2}\sigma ^2 \;dt+\frac{\partial F}{
\partial S} r \;dt +\frac{\partial F}{\partial t} dt = r F \;dt
\end{equation}

This is a simple derivation that did not need the hedging argument using self-financing portfolios. But this approach is useful for the following examples. 

###EXAMPLE

Assume that $X$ follows an arithmetic Brownian motion. We start by defining a contingent claim $F(X)$, which is a traded asset that earns $X \;dt$ continuously to infinity. What is the price today of this claim?

The specifications are as follows:

\begin{eqnarray*}
dX &=&\alpha Xdt+\sigma XdZ \\
F(X,t) &=&F(X) 
\end{eqnarray*}

We specify the change in the portfolio containing $F(X)$ as follows:

\begin{equation}
dF =\left[ \alpha XF_x+\frac 12\sigma ^2X^2F_{xx}\right] dt+\sigma XF_xdZ
\end{equation}

$F$ also changes by $X\; dt$. So the expected change in value of the portfolio is:

\begin{eqnarray*}
E\left[ dF\right] +E[Xdt] &=&\left[ \alpha XF_x+\frac 12\sigma ^2X^2F_{xx}+X%
\right] dt 
\end{eqnarray*}

The risk-neutral pricing approach is to take this expected value and set it equal to the risk free return on the portfolio: 

\begin{eqnarray*}
\left[ \alpha XF_x+\frac 12\sigma ^2X^2F_{xx}+X%
\right] dt &=&rFdt \\
\alpha XF_x+\frac 12\sigma ^2X^2F_{xx}+X-rF &=&0
\end{eqnarray*}


Guess a solution: $F(X)=\beta X$.

\begin{eqnarray*}
0 &=&\beta \alpha X+X-r\beta X \\
-1 &=&\beta (\alpha -r) \\
\beta &=&\frac 1{r-\alpha }
\end{eqnarray*}

Therefore:

\begin{equation}
F(X)=\frac{X}{r-\alpha} 
\end{equation}

Note: $\sigma$ does not enter the solution - see the parallel to the annuity formula. 


###Example

Supppose $F(X)$ is a perpetual contract that pays $X$ in exchange for $F$, if the Poisson event occurs. What is the price of this contract today? Assume that $X$ follows a jump-diffusion process

\begin{equation}
dX =\alpha Xdt+\sigma XdZ+JdQ
\end{equation}

$dF$ follows the equation

\begin{eqnarray*}
dF &=&\left[ \alpha XF_x+\frac 12\sigma ^2X^2F_{xx}\right] dt
+\lambda \left[
X-F\right] dQ+\frac 12\sigma XdZF_x 
\end{eqnarray*}

Using the risk-neutral pricing approach, we have:

\begin{eqnarray*}
E\left[ dF\right] &=&\left[ \alpha XF_x+\frac 12\sigma ^2X^2F_{xx}+\lambda %
\left[ X-F\right] \right] dt =rFdt \\
(r+\lambda )F &=&\alpha XF_x+\frac 12\sigma ^2X^2F_{xx}+\lambda X
\end{eqnarray*}

Guess a solution: $F(X)=\beta X$.

\begin{eqnarray*}
(r+\lambda )\beta X &=&\alpha X\beta +0+\lambda X \\
\lambda &=&[(r+\lambda )-\alpha ]\beta \\
\beta &=&\frac \lambda {r+\lambda -\alpha }
\end{eqnarray*}

Therefore:

\begin{equation}
F(X)=\left( \frac \lambda {r+\lambda -\alpha }\right) X 
\end{equation}

What happens is $\lambda \rightarrow \infty$?


###Example

Same problem as above but add a jump which is the total loss of $F$. What is the price today?
\begin{eqnarray*}
dX &=&\alpha Xdt+\sigma XdZ+JdQ \\
J &=&-F \\
dF &=&\left[ \alpha XF_x+\frac 12\sigma ^2X^2F_{xx}+X\right] dt+\sigma
XF_xdZ+[0-F]dQ \\
E\left[ dF\right] &=&\left[ \alpha XF_x+\frac 12\sigma ^2X^2F_{xx}+X-\lambda
F\right] dt \\
&=&rFdt \\
0 &=&\alpha XF_x+\frac 12\sigma ^2X^2F_{xx}+X-(r+\lambda )F
\end{eqnarray*}
Solution:
\begin{eqnarray*}
F(X) &=&\frac X{r+\lambda -\alpha } \\
&=&\frac X{r-(\alpha -\lambda )}
\end{eqnarray*}
Note: this is just the default adjusted growth rate of perpetuity.


###Black-Scholes-Merton with Jump to Default

Assume the usual **risk-neutral** GBM for the stock but with a possible jump to default, when the stock becomes worthless. 

\begin{equation}
dS = rS \;dt + \sigma S \;dB - S dQ(\lambda)
\end{equation}

Given an option $F(S,t)$ the usual approach gives:

\begin{equation}
dF = \frac 12\frac{\partial ^2F}{\partial S^2}\sigma ^2dt+\frac{\partial F}{
\partial S}\left[ r dt+\sigma dZ\right] +\frac{\partial F}{\partial t} dt - F dQ 
\end{equation}

Using the risk-neutral approach we derive the PDE:

\begin{equation}
E(dF) = \frac 12\frac{\partial ^2F}{\partial S^2}\sigma ^2dt+\frac{\partial F}{
\partial S} r dt +\frac{\partial F}{\partial t} dt - \lambda F dt = rF \;dt
\end{equation}

Which gives the usual BS PDE: 

\begin{equation}
\frac 12\frac{\partial ^2F}{\partial S^2}\sigma ^2 \;dt+\frac{\partial F}{
\partial S} r \;dt +\frac{\partial F}{\partial t} dt = (r + \lambda) F \;dt
\end{equation}

Solving this PDE in the usual fashion gives the solution, very similar to Black-Scholes, i.e., for a call option with maturity $T$

\begin{eqnarray}
F(S,t) &=& S N(d_1) - K e^{-R(T-t)} N(d_2) \\
d_1 &=& \frac{\ln(S/K) + (R+\frac{1}{2}\sigma^2)(T-t)}{\sigma \sqrt{T-t}} \\
d_2 &=& d_1 - \sigma \sqrt{T-t} \\
R &=& r + \lambda
\end{eqnarray}

In other words, the same equation as before, but with an adjusted interest rate. 

How will the call price be different for this firm where default is possible versus a firm where no default is possible, i.e., $\lambda=0$? Is this strange?

###Gaussian Jumps Model

See Merton 1976, **Continuous Time Finance**, Chapter 9.

\begin{equation}
\frac{dS}S=(\alpha -\lambda k)dt+\sigma dZ+JdQ(\lambda ) 
\end{equation}

$dZ$, $dQ$ independent $(dZ\;dQ=0)$. $J=Y-1=\% \Delta$ in stock price; $S$ becomes $S(1+\% \Delta)$. Note that $k = E[Y-1]$. 

\begin{eqnarray*}
\frac{dS}S &=&(\alpha -\lambda k)dt+\sigma dZ, \quad\mbox{ if no jump} \\
\frac{dS}S &=&(\alpha -\lambda k)dt+\sigma dZ+(Y-1), \quad \mbox{ if jump} \\
\frac{S(t)}{S_0} &=&\left\{ \exp \left[ (\alpha -\frac{\sigma ^2}2-\lambda
k)dt+\sigma Z(t)\right] \right\} Y(n) \\
Y(n) &=&\prod\limits_{j=1}^nY_j
\end{eqnarray*}

$\frac{dS}S=Y-1$; $dS=SY-S$; $S_{t+dt}-S_t=S_tY-S_t$; $S_{t+dt}=S_tY$; diffusion goes first.

Option: $\equiv F(S,t)$;  Apply Ito's Lemma.

\begin{eqnarray*}
dF&=&\frac 12F_{ss}\sigma ^2 S^2 \;dt+F_s\left[ (\alpha -\lambda k)\; S\; dt+\sigma S \; dZ \right] \\
&&+F_t\; dt+\left[ F(SY,t)-F(S,t)\right] \; dQ 
\end{eqnarray*}

Use risk-neutral arguments: set $\alpha =r$, $E[dF]=(r+\lambda )F\; dt$. Set $\lambda =0$, the jump premium is zero. 

\begin{equation}
0=\frac 12F_{ss}\sigma ^2S^2+F_s(r-\lambda k)S+F_t+\lambda E\left[ F(SY)-F(S)\right] -rF 
\end{equation}

Define $X_n$ to be the product of $n$ iid random variables of type $Y$. Let $Y$ be lognormal with $\ln(Y) \sim N(\mu ,\gamma ^2)$.

\begin{equation}
X_n=\prod\limits_{j=1}^nY_j 
\end{equation}

Do risk-neutral valuation by integration, simplify:

\begin{eqnarray*}
S_t &=&S_0\exp \left[ X+\sum\limits_{j=1}^Qv_j\right] \\
&&Q\mbox{ is a Poisson variable} \\
X &\sim &N\left[ \left( r-\frac 12\sigma ^2-\lambda k\right) t,\sigma ^2t%
\right] \\
v_j &\sim &N[\mu ,\gamma ^2] \\
E[X] &=&\left[ r-\frac 12\sigma ^2-\lambda \left( e^{\mu +\frac 12\gamma
^2}-1\right) \right] t
\end{eqnarray*}

Given $n$ jumps exactly

\begin{eqnarray*}
Y_n &=&X+\sum\limits_{j=1}^n v_j \\
E[Y_n] &=&\left[ r-\frac 12\sigma ^2-\lambda \left( e^{\mu +\frac 12\gamma
^2}-1\right) +\frac{n\mu }t\right] t \\
Var[Y_n] &=&\left[ \sigma ^2+\frac{n\gamma ^2}t\right] t
\end{eqnarray*}

Then, a call option at time zero for maturity $t$ is:

\begin{eqnarray*}
&& e^{-rt}\; E\left[ \max (S_{t}-K,0)\right]  \\
&=& e^{-rt} \; \sum\limits_{n=0}^{\infty }\frac{%
e^{-\lambda t}(\lambda t)^{n}}{n!}E\left\{ \max (S_{0}e^{Y_{n}}-K,0)\right\} \\
&=&\sum\limits_{n=0}^{\infty }\frac{e^{-\overline{\lambda }t}(\overline{%
\lambda }t)^{n}}{n!}\times  \\
&&\mbox{BS-call}\left[ S_{0};t;K;r+\frac{1}{t}(\mu +\frac{1}{2}\gamma
^{2})-\lambda (e^{\mu +\frac{1}{2}\gamma ^{2}}-1);\sigma ^{2}+n\gamma ^{2}%
\right] 
\end{eqnarray*}

where $\overline{\lambda }=\lambda e^{\mu +\frac 12\gamma ^2}$. This
derivation is left to the reader. Also show that if the jump is tredless
(i.e. $\mu =\frac{-\gamma ^2}2$) then the call formula is given by

\begin{equation}
Call=\sum\limits_{n=0}^\infty \frac{e^{-\lambda t}(\lambda t)^n}{n!}\mbox{
Call}\left[ S_0;t;K;r;\sigma ^2+\frac{n\gamma ^2}t\right] 
\end{equation}


###SIMULATING THE JUMP-DIFFUSION PROCESS

From the preceding discussion note that the process for the stock price may be written as:

\begin{eqnarray*}
S_t &=&S_0\exp \left[ X+\sum\limits_{j=1}^Qv_j\right] \\
&&Q\mbox{ is a Poisson variable} \\
X &\sim &N\left[ \left( r-\frac 12\sigma ^2-\lambda k\right) t,\sigma ^2t%
\right] \\
v_j &\sim &N[\mu ,\gamma ^2] \\
Y &=& e^{v_j} \\
k &=& E(Y-1) = e^{\mu +\frac 12\gamma^2}-1\\
E[X] &=&\left[ r-\frac 12\sigma ^2-\lambda \left( e^{\mu +\frac 12\gamma^2}-1\right) \right] t
\end{eqnarray*}

These equations may then be used for simulating the jump-diffusion process, next. 

```python
from scipy.stats import poisson

#PARAMETERS
m = 1000
n = 252
s0 = 100.0
r = 0.02
sigma = 0.20
T = 1.0
mu = -0.10
gamma = 0.50
lam = 1  #one jump a year

#SET UP
s = zeros((m,n+1))
h = T/n
k = exp(mu+0.5*gamma**2)-1

#SIMULATE
s[:,0] = s0
for j in range(1,n+1):
    Q = poisson.rvs(lam*h, size=m)        
    s[:,j] = s[:,j-1]*exp((r-0.5*sigma**2-lam*k)*h + sigma*sqrt(h)*randn(m) + (Q*mu + sqrt(Q)*gamma*randn(m)))
    
s.shape
```

```python
#PRICE CALL
strike = 101.0
call_value = exp(-r*T)*mean(maximum(0,s[:,n]-strike))
print("Call = ",call_value)
```

```python
#BLACK-SCHOLES-MERTON MODEL
from scipy.stats import norm
def BSM(S,K,T,sig,rf,dv,cp):  #cp = {+1.0 (calls), -1.0 (puts)}
    d1 = (log(S/K)+(rf-dv+0.5*sig**2)*T)/(sig*sqrt(T))
    d2 = d1 - sig*sqrt(T)
    return cp*S*exp(-dv*T)*norm.cdf(d1*cp) - cp*K*exp(-rf*T)*norm.cdf(d2*cp)
```

```python
#IMPLIED VOLATILITY
def BSMvol(sig,S,K,T,rf,dv,cp,optprice):
    return BSM(S,K,T,sig,rf,dv,cp)-optprice

from scipy.optimize import fsolve

dv=0.0; cp=1.0

sol = fsolve(BSMvol,0.12,args=(s0,strike,T,r,dv,cp,call_value))
print("implied vol = ",sol)
```

```python
strikes = arange(75.0,130.0,5.0)
len_k = len(strikes)
impvol = zeros(len_k)
j=0
for strike in strikes:
    call_value = exp(-r*T)*mean(maximum(0,s[:,n]-strike))
    sol = fsolve(BSMvol,0.12,args=(s0,strike,T,r,dv,cp,call_value))
    impvol[j] = sol
    print("Call = ",call_value,"implied vol = ",sol)
    j = j + 1
    
plot(strikes,impvol)
grid()
```
